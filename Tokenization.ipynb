## Stemming (м–ҙк°„ м¶”м¶ң)
-  REмқ„ мӮ¬мҡ©н•ҳм—¬ мҲҳмһ‘м—…мңјлЎң н•  мҲҳлҸ„ мһҲмңјлӮҳ, NLTK лӮҙмһҘ stemmerлҠ” л¶Ҳк·ңм№ҷм Ғмқё кІҪмҡ°л“Өмқ„ мІҳлҰ¬н•  мҲҳ мһҲмңјлҜҖлЎң лҚ” нҡЁмңЁм ҒмқҙлӢӨ.
- NLTK лӮҙмһҘ stemmer:
  - Porter stemmer
  - Lancaster stemmer

````python
porter = nltk.PorterStemmer()
print([porter.stem(t) for t in tokens])
````

````python
lancaster = nltk.LancasterStemmer()
print([lancaster.stem(t) for t in tokens])
````

- StemmingмқҖ мһҳ м •мқҳлҗң кіјм •мқҙ м•„лӢҲл©° лӘ©м Ғм—җ л§һлҠ” кІғмқ„ м„ нғқн•ҙ мӮ¬мҡ©н•ңлӢӨ
- н…ҚмҠӨнҠё мқён…ҚмӢұ л“ұ мқјл°ҳм Ғмқё NLP мһ‘м—…м—җ мһҲм–ҙ Porter stemmer м„ұлҠҘмқҙ Lancaster stemmer ліҙлӢӨ мўӢмқҢ

#### StemmerлҘј мӮ¬мҡ©н•ң н…ҚмҠӨнҠё мқёлҚұмӢұ
- `concordance` viewлҠ” мЈјм–ҙм§„ лӢЁм–ҙк°Җ н…ҚмҠӨнҠё лӮҙм—җм„ң м¶ңнҳ„н•ҳлҠ” кІғмқ„ мЈјліҖ лӢЁм–ҙл“Ө(context)кіј н•Ёк»ҳ ліҙм—¬мӨҖлӢӨ.

````python
class IndexedText(object):

    def __init__(self, stemmer, text):
        self._text = text
        self._stemmer = stemmer
        self._index = nltk.Index((self._stem(word), i) # nltk.Index source code: https://tedboy.github.io/nlps/_modules/nltk/util.html#Index
                                 for (i, word) in enumerate(text))
    def concordance(self, word, width=40):
        key = self._stem(word)
        wc = int(width/4)                # words of context, мҷң 4лЎң лӮҳлҲҙмқ„к№Ң?
        for i in self._index[key]:
            lcontext = ' '.join(self._text[i-wc:i])
            rcontext = ' '.join(self._text[i:i+wc])
            ldisplay = '{:>{width}}'.format(lcontext[-width:], width=width)  # ':>{width}' width кёёмқҙ л§ҢнҒј н‘ңмӢңн•ҳлҗҳ мҳӨлҘёмӘҪ л§һм¶Өн•ңлӢӨ
            rdisplay = '{:{width}}'.format(rcontext[:width], width=width)
            print(ldisplay, rdisplay)

    def _stem(self, word):
        return self._stemmer.stem(word).lower()
````

````python
import nltk
nltk.download('webtext')
````

````python
porter = nltk.PorterStemmer()
grail = nltk.corpus.webtext.words('grail.txt')
text = IndexedText(porter, grail)
text.concordance('lie')
````

### лӢЁм–ҙ м–ҙк°„ м°ҫкё°
- кІҖмғү м—”м§„м—җ мһ…л Ҙн•  л•Ң лӢЁм–ҙмқҳ лҒқл¶Җл¶„мқҙ лӢӨлҘё кІғмқ„ мӢ кІҪм“°м§Җ м•Ҡм•„лҸ„ лҗҳлҠ” кІғмқҖ stemming кё°лҠҘ л•Ңл¬ёмқҙлӢӨ.
- к°ҖмһҘ к°„лӢЁн•ң л°©лІ•мңјлЎң suffixмІҳлҹј ліҙмқҙлҠ” кІғмқҖ лӢӨ мһҳлқјлІ„лҰ¬лҠ” л°©лІ•мқ„ мӮ¬мҡ©н•ҙліҙмһҗ.
- л¬јлЎ  NLTKм—җлҠ” stemmerк°Җ мһҲлӢӨ.

````python
def stem(word):
    for suffix in ['ing', 'ly', 'ed', 'ious', 'ies', 'ive', 'es', 's', 'ment']:
        if word.endswith(suffix):
            return word[:-len(suffix)]
    return word
````

- disjunction('OR'ing) of all suffixes
- disjunctionмқҳ лІ”мң„лҘј н•ңм •н•ҳкё° мң„н•ҙ кҙ„нҳёлҘј м“ҙлӢӨ.

````python
re.findall(r'^.*(ing|ly|ed|ious|ies|ive|es|s|ment)$', 'processing')
````

- лӢЁм–ҙлҘј м¶ңл Ҙн•ҳл Өл©ҙ `?:` лҘј м¶”к°Җн•ҙм•ј н•ңлӢӨ.

- `*` м—°мӮ°мһҗмқҳ 'non-greedy' лІ„м „ `*?`мқ„ мӮ¬мҡ©н•ҳмһҗ.

````python
text = "doyouseethekittyseethedoggydoyoulikethekittylikethedoggy"
seg1 = "0000000000000001000000000010000000000000000100000000000"  # мҙҲкё° segmentation
seg2 = "0100100100100001001001000010100100010010000100010010000"  # лӘ©н‘ң segmentation
````

````python
def segment(text, segs):
    words = []
    last = 0
    for i in range(len(segs)):
        if segs[i] == '1':
            words.append(text[last:i+1])
            last = i+1
    words.append(text[last:])
    return words
````

- learnerлҠ” лӢЁм–ҙлҘј м·Ёл“қн•ҳм—¬ лӮҙл¶Җ lexiconм—җ м ҖмһҘн•ңлӢӨ.
- [Brent (1995)](https://www.nltk.org/book/bibliography.html#brent1995)мқҳ л°©лІ•мқ„ л”°лқј **лӘ©м Ғн•ЁмҲҳ**лҘј л§Ңл“Ө мҲҳ мһҲлӢӨ(к·ёлҰј 3.8)
- лӘ©м Ғн•ЁмҲҳ = мӮ¬м „(lexicon) нҒ¬кё°(лӢЁм–ҙ лӮҙ кёҖмһҗ мҲҳ + к°Ғ лӢЁм–ҙмқҳ лҒқмқ„ лӮҳнғҖлӮҙлҠ” лі„лҸ„мқҳ кө¬нҡҚ л¬ёмһҗ)мҷҖ lexiconм—җм„ң мӣҗлһҳ н…ҚмҠӨнҠёлҘј ліөмӣҗн•ҳкё° мң„н•ҙ н•„мҡ”н•ң м •ліҙлҹүмқ„ кё°л°ҳмңјлЎң н•ҳлҠ” scoring function

<img src="https://github.com/kikim6114/nlp2026/blob/main/images/segmentation_objective.png?raw=1" width="600" style="margin-left: auto; margin-right: auto">
<p style="text-align: center;">м„ёнҒ¬л©ҳн…Ңмқҙм…ҳ лӘ©м Ғн•ЁмҲҳ</p>

````python
def evaluate(text, segs):
    words = segment(text, segs)
    text_size = len(words)
    lexicon_size = sum(len(word) + 1 for word in set(words))
    return text_size + lexicon_size
````

- л§Ҳм§Җл§ү лӢЁкі„лҠ” лӘ©м Ғн•ЁмҲҳлҘј мөңмҶҢнҷ”мӢңнӮӨлҠ” 0кіј 1мқҳ нҢЁн„ҙмқ„ м°ҫлҠ” кІғмқҙлӢӨ.
- мөңм Ғнҷ”лҗң segmentationм—җ лӢЁм–ҙ *thekitty*к°Җ лӮҳмҳӨлҠ” мқҙмң лҠ” мқҙкІғмқ„ лҚ” л¶„лҰ¬н•ҳкё°м—җлҠ” м •ліҙк°Җ л¶ҖмЎұн•ҳкё° л•Ңл¬ё

**Non-Deterministic Search Using Simulated Annealing:**  
- Phrase segmenationм—җл§Ң м Ғмҡ©
- "temperature"м—җ л№„лЎҖн•ҳм—¬ 0кіј 1мқ„ л¬ҙмһ‘мң„лЎң көҗлһҖмӢңнӮЁлӢӨ
- л§Ө л°ҳліөл§ҲлӢӨ temperatureлҠ” лӮҙл Өк°Җкі  лӢЁм–ҙ кІҪкі„мқҳ көҗлһҖмқҖ мӨ„м–ҙл“ лӢӨ
- м•Ңкі лҰ¬мҰҳмқҙ non-deterministicмқҙлҜҖлЎң кІ°кіјлҠ” л§ӨлІҲ лӢ¬лқјм§Ҳ мҲҳ мһҲлӢӨ

````python
from random import randint

def flip(segs, pos):
    return segs[:pos] + str(1-int(segs[pos])) + segs[pos+1:]

def flip_n(segs, n):
    for i in range(n):
        segs = flip(segs, randint(0, len(segs)-1))
    return segs

def anneal(text, segs, iterations, cooling_rate):
    temperature = float(len(segs))
    while temperature > 0.5:
        best_segs, best = segs, evaluate(text, segs)
        for i in range(iterations):
            guess = flip_n(segs, round(temperature))
            score = evaluate(text, guess)
            if score < best:
                best, best_segs = score, guess
        score, segs = best, best_segs
        temperature = temperature / cooling_rate
        print(evaluate(text, segs), segment(text, segs))
    print()
    return segs
````

## 1) dictionary мӮ¬мҡ©н•ҳкё°

````python
from nltk.tokenize import sent_tokenize
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
````

````python
raw_text = "A barber is a person. a barber is good person. a barber is huge person. he Knew A Secret! The Secret He Kept is huge secret. Huge secret. His barber kept his word. a barber kept his word. His barber kept his secret. But keeping and keeping such a huge secret to himself was driving the barber crazy. the barber went up a huge mountain."
````

````python
# л¬ёмһҘ л¶„н• 
sentences = sent_tokenize(raw_text)
sentences
````

````python
vocab = {}
preprocessed_sentences = []
stop_words = set(stopwords.words('english'))

for sentence in sentences:
    # лӢЁм–ҙ нҶ нҒ°нҷ”
    tokenized_sentence = word_tokenize(sentence)
    result = []

    for word in tokenized_sentence:
        word = word.lower() # лӘЁл“  лӢЁм–ҙлҘј мҶҢл¬ёмһҗнҷ”н•ҳм—¬ лӢЁм–ҙмқҳ к°ңмҲҳлҘј мӨ„мқёлӢӨ.
        if word not in stop_words: # лӢЁм–ҙ нҶ нҒ°нҷ” лҗң кІ°кіјм—җ лҢҖн•ҙм„ң л¶Ҳмҡ©м–ҙлҘј м ңкұ°н•ңлӢӨ.
            if len(word) > 2: # лӢЁм–ҙ кёёмқҙк°Җ 2мқҙн•ҳмқё кІҪмҡ°м—җ лҢҖн•ҳм—¬ м¶”к°ҖлЎң лӢЁм–ҙлҘј м ңкұ°н•ңлӢӨ.
                result.append(word)
                if word not in vocab:
                    vocab[word] = 0
                vocab[word] += 1
    preprocessed_sentences.append(result)
print(preprocessed_sentences)
````

````python
print(vocab)
````

````python
print(vocab["barber"]) # 'barber'лқјлҠ” лӢЁм–ҙмқҳ л№ҲлҸ„мҲҳ м¶ңл Ҙ
````

````python
vocab_sorted = sorted(vocab.items(), key = lambda x:x[1], reverse = True)
print(vocab_sorted)
````

````python
word_to_index = {}
i = 0
for (word, frequency) in vocab_sorted :
    if frequency > 1 : # л№ҲлҸ„мҲҳк°Җ мһ‘мқҖ лӢЁм–ҙлҠ” м ңмҷё.
        i = i + 1
        word_to_index[word] = i

print(word_to_index)
````

````python
vocab_size = 5
words_frequency = [word for word, index in word_to_index.items() if index >= vocab_size + 1] # мқёлҚұмҠӨк°Җ 5 мҙҲкіјмқё лӢЁм–ҙ м ңкұ°
for w in words_frequency:
    del word_to_index[w] # н•ҙлӢ№ лӢЁм–ҙм—җ лҢҖн•ң мқёлҚұмҠӨ м •ліҙлҘј мӮӯм ң
print(word_to_index)
````

````python
word_to_index['OOV'] = len(word_to_index) + 1
````

````python
encoded_sentences = []
for sentence in preprocessed_sentences:
    encoded_sentence = []
    for word in sentence:
        try:
            encoded_sentence.append(word_to_index[word])
        except KeyError:
            encoded_sentence.append(word_to_index['OOV'])
    encoded_sentences.append(encoded_sentence)
print(encoded_sentences)
````
